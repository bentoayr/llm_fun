## Main function

```python
import string
import openai
from openai import OpenAI # Import the new OpenAI client

# Initialize the OpenAI client globally or within the function if preferred.
# For simplicity, we'll initialize it once here. Make sure openai.api_key is set.
client = OpenAI(api_key=openai.api_key)

def call_llm_with_prompt_template(prompt_template: str, input_prompt: str) -> str:
    """Calls an LLM with a given prompt template and input.

    Args:
        prompt_template: A string containing a placeholder `$input`.
        input_prompt: The string to substitute into the template.

    Returns:
        The LLM's response.
    """
    # Use string.Template for substitution
    full_prompt = string.Template(prompt_template).substitute(input=input_prompt)

    try:
        response = client.chat.completions.create( # Use the new client.chat.completions.create
            model=openai_model_name,
            messages=[{"role": "user", "content": full_prompt}]
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        return f"Error calling LLM: {e}"

print("Function `call_llm_with_prompt_template` defined.")
```

### Simple Test


```python
prompt1 = "Summarize the following text in one sentence: $input"
prompt2 = "Mathematica is a symbolic computation system developed by Wolfram."

result = call_llm_with_prompt_template(prompt1, prompt2)
print(result)
```

### Some Fun



```python
prompt_for_improvement = "Improve the following prompt: `$input`. The output should just have the improved prompt. Nothing else. No comments, nothing."

current_prompt_to_refine = prompt_for_improvement # Start with the prompt itself

print(f"Initial prompt for refinement: {current_prompt_to_refine}\n")

for i in range(1, 11):
    print(f"--- Iteration {i} ---")
    refined_prompt = call_llm_with_prompt_template(prompt_for_improvement, current_prompt_to_refine)
    # Basic cleaning: remove extra quotes or backticks if the LLM adds them
    refined_prompt = refined_prompt.strip().strip('`').strip('"').strip()

    print(refined_prompt)
    current_prompt_to_refine = refined_prompt # Use the refined prompt for the next iteration
    print("\n")
```
